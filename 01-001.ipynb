```python
# 파이썬에서 지도를 다루기 위한 핵심 도구인 GeoPandas를 불러옵니다.
# gpd라는 별명으로 부르겠다는 뜻입니다.
import geopandas as gpd
import matplotlib.pyplot as plt
print("라이브러리 불러오기 성공!")
```

```python
# 내 컴퓨터에 있는 지적도 파일(SHP)을 읽어옵니다.
# 경로(path)는 파일이 있는 실제 위치로 수정해야 합니다.
# 윈도우에서는 경로에 역슬래시(\) 대신 슬래시(/)를 쓰는 게 오류가 적습니다.

#file_path = '/source/01_shp/LSMD_CONT_LDREG_5174_41310_202602.shp'  # 예시 파일 경로
file_path = r'F:\src_py\01\LSMD_CONT_LDREG_5174_경기_구리시\LSMD_CONT_LDREG_5174_41310_202602.shp'  # 예시 파일 경로

# read_file 명령어로 데이터를 읽어서 'gdf'라는 변수(그릇)에 담습니다.
gdf = gpd.read_file(file_path)
print("1.SHP 파일읽기")
# 데이터가 잘 들어왔는지 상위 5개만 눈으로 확인해봅니다.
print("데이터 로딩 완료. 총 필지 수:", len(gdf))
gdf.head()
print("2.SHP 헤드읽기")
```

```python
# 드디어 지도를 그립니다.
# figsize는 그림의 크기(가로, 세로)를 의미합니다.

gdf.plot(figsize=(10, 5))

# 그래프 제목 달기
plt.title("My First Cadastral Map with Python")
plt.show()
```
